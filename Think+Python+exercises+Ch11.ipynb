```python
'''Exercise 11.1. Write a function that reads the words in words.txt and stores them as keys in a
dictionary. It doesn’t matter what the values are. Then you can use the in operator as a fast way to
check whether a string is in the dictionary.'''

def read_dic():
    d = {}
    fin = open('words.txt')
    for line in fin:
        word = line.strip()
        d[word] = word
    return d

read_dic()
```

```python
# original function
def invert_dict(d):
    inverse = dict()
    for key in d:
        val = d[key]
        if val not in inverse:
            inverse[val] = [key]
        else:
            inverse[val].append(key)
    return inverse

d = {'a': 1, 'p': 1, 'r': 2, 't': 1, 'o': 1}
invert_dict(d)
```

```python
# variation on original function, using defaultdict 
from collections import defaultdict

def inv_d(d):
    inverse = defaultdict(list)
    for key in d:
        val = d[key]
        inverse[val].append(key)
    return inverse

inv_d(d)
```

```python
'''Exercise 11.4. If you did Exercise 10.7, you already have a function named has_duplicates that
takes a list as a parameter and returns True if there is any object that appears more than once in the
list. Use a dictionary to write a faster, simpler version of has_duplicates.'''

# original function
def has_duplicates(a):
    if len(a) != len(set(a)):
        return True
    else:
        return False
    
def has_dup(a):
    d = dict()
    for c in a:
        if c not in d:
            d[c] = 1
        else:
            d[c] += 1
    if len(a) == len(d):
        return False
    else:
        return True
        

a = [1,2,3,3]        
has_dup(a)
```

```python
def has_duplicate(t):
    """Checks whether any element appears more than once in a sequence.

    Simple version using a for loop.

    t: sequence
    """
    d = {}
    for x in t:
        if x in d:
            return True
        d[x] = True
    return False

b = [1,2,3,4,4]
has_duplicate(b)
```

```python
'''Exercise 11.5. Two words are “rotate pairs” if you can rotate one of them and get the other (see
rotate_word in Exercise 8.5). Write a program that reads a wordlist and finds all the rotate pairs.'''

def read_dic():
    d = {}
    fin = open('words.txt')
    for line in fin:
        word = line.strip()
        if word == word[::-1]:
            d[word] = word[::-1]
        else:
            pass        
    return d

read_dic()
```
